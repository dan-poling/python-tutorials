```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
# compare population growth between US & China
```

```python
# This shows data when [this] is True
data[data.country == 'United States']
```

```python
us = data[data.country == 'United States']
```

```python
# This sets china to data when [this] is True
china = data[data.country == "China"]
```

```python
china
```

```python
# plot the US information
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.title("Population Growth, US / China")
plt.xlabel("Year")
plt.ylabel('Population (Mil)')
plt.legend(['US', "China"])
plt.show
```

```python
# Percentage Growth with 1st year normalize in cell above
# Normalize: population / base year * 100 for percent
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.title("Percentage Population Growth, US / China")
plt.xlabel("Year")
plt.ylabel('Population Growth')
plt.legend(['US', "China"])
plt.show
```
